# LONG URL- BITSY URL

## Business Problem:

Generally URLs are very long to read and write also. It take lot of time to type a complete URL during heavy workload.

Sometimes, they can get a little complex (especially when you’re tracking them).
Sometimes, they can be a bit distracting.

And if you’re trying to share helpful information with a growing audience, then you don’t want the links to all that golden content to upstage your efforts!

## Solution:

Bitsy URL. BitsyURL is a simple solution to shorten a very long, boring, complex and distracting URL. Here you can convert long URL in a very bitsy URL in a single click which is very easy to read, write and share with your customers or friends.


```python
from tkinter import *
import random
import string


root=Tk()
root.geometry('500x500')
root.title('BitsyURL.com')
root.configure(background='#03254c')

# Empty string is created to store BitsyURL
empty_dict=dict()

# A function get_BitsyURL is defined to which we will pass longURL and function will return shortURL for that particular longURL
def get_BitsyURL(longURL):
    
    length= random.randint(4,8) # Here we will generate random length of our shortURL
    
    characters= string.ascii_lowercase
    
    BitsyURL= ''.join(random.choice(characters) for i in range(length)) # shortURL of length 'length' is generated
    
    if BitsyURL in empty_dict:
        return get_BitsyURL(longURL)
    else:
        empty_dict[BitsyURL]=longURL
        
    website="https://www.BitsyURL.com/" + BitsyURL
    bitsyURL.config(text=website)
    
    
def get_longURL(BitsyURL):
    
    k= BitsyURL[25:]
    
    if k in empty_dict:
        return empty_dict[k]
    else:
        return None

my_label=Label(root, text='Enter URL to Shorten', font=('verdana',16)) 
my_label.configure(background='#03254c', foreground='white')
my_label.pack(pady=20)

my_entry=Entry(root, font=('verdana',16))
my_entry.configure(width=35)
my_entry.pack(pady=20)   

my_button=Button(root, text='Make BitsyURL', command=lambda:get_BitsyURL(my_entry), font=('verdana',16))
my_button.configure(background='grey', foreground='white')
my_button.pack(pady=20)

bitsyURL_label=Label(root, text='Here is your BitsyURL', font=('verdana',16)) 
bitsyURL_label.configure(background='#03254c', foreground='white', width=35)
bitsyURL_label.pack(pady=30)

bitsyURL=Label(root, width=35, height=1,font=('verdana',16))
bitsyURL.pack(pady=10)


root.mainloop()
```
